```python
TrainTx = pd.read_csv('../input/type-data-nbaiot-processed/TrainTx.csv')
TrainTy = pd.read_csv('../input/type-data-nbaiot-processed/TrainTy.csv')
ValTx = pd.read_csv('../input/type-data-nbaiot-processed/ValTx.csv')
ValTy = pd.read_csv('../input/type-data-nbaiot-processed/ValTy.csv')
```

```python
TotalTrainDx = pd.concat([TrainTx, ValTx], axis=0) 
TotalTrainDy = pd.concat([TrainTy, ValTy], axis=0)
```

```python
params = {
    'learning_rate':[0.10, 0.20, 0.30],
    'max_depth':[3 ,5, 7],
    'min_child_weight':[3,5,7],
    'gamma':[0.1, 0.2, 0.3],
    'alpha':[0.01, 0.03, 0.06],
    'n_estimators': [20,40, 60, 80]
}
```

```python
xg = xgb.XGBClassifier(objective ='multi:softmax' num_class=11)
```

```python
rf_grid = RandomizedSearchCV(estimator = xg, 
                       param_distributions = params,
                       verbose=2,
                       n_jobs = -1,
                       cv = 5, n_iter = 50)
```

```python
rf_grid.fit(TrainTx, TrainTy)
```

```python
best_model = rf_grid.best_estimator_
```

```python
y_pred = best_model.predict(ValTx)
```

```python
accuracy = accuracy_score(ValTy, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
TestTx = pd.read_csv('../input/type-data-nbaiot-processed/TestTx.csv')
TestTy = pd.read_csv('../input/type-data-nbaiot-processed/TestTy.csv')
```

```python
TestPreds = best_model.predict(TestTx)
metrics.accuracy_score(TestTy, TestPreds)
```

## Accuracies

```python
def tellAcc(x, y):
    return metrics.accuracy_score(y, best_model.predict(x))
```

## Confusion Matrix

```python
cm = pd.DataFrame(confusion_matrix(TestPreds, TestTy, labels=range(11)))
cm
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(cm, annot=True, cmap='Blues')
plt.xlabel('Actual')
plt.ylabel('Predicted');
```

```python
def recall(type, TestPreds, TestDy, labels):
    confMat = confusion_matrix(TestPreds, TestDy, labels=labels)
    true = confMat[type][type]
    total = 0
    for i in range(len(labels)):
        total += confMat[i][type]
    return true/total

def precision(type, TestPreds, TestDy, labels):
    confMat = confusion_matrix(TestPreds, TestDy, labels=labels)
    true = confMat[type][type]
    total = 0
    for i in range(len(labels)):
        total += confMat[type][i]
    return true/total

def specificity(type, TestPreds, TestDy, labels):
    confMat = confusion_matrix(TestPreds, TestDy, labels=labels)
    true = 0
    for i in range(len(labels)):
        if(i==type):
            continue;
        true += confMat[i][i]
    total = 0
    for j in range(len(labels)):
        if(j==type):
            continue
        for i in range(len(labels)):
            total += confMat[i][j]
    return true/total
```

```python
labels=range(11)
```

```python
for i in range(len(labels)):
    print('recall of class '+str(i)+': '+str(recall(i, TestPreds, TestTy, labels)))
```

```python
for i in range(len(labels)):
    print('precision of class '+str(i)+': '+str(precision(i, TestPreds, TestTy, labels)))
```

```python
for i in range(len(labels)):
    print('specificity of class '+str(i)+': '+str(specificity(i, TestPreds, TestTy, labels)))
```
