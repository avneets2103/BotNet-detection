```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import warnings
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
warnings.filterwarnings('ignore')
```

```python
list_of_files = os.listdir('../input/nbaiot-dataset')
len(list_of_files)
```

```python
# catergorical data
def type(x):
    if 'benign' in x:
        return 0
    elif 'gafgyt.combo' in x:
        return 1
    elif 'gafgyt.junk' in x:
        return 2
    elif 'gafgyt.scan' in x:
        return 3
    elif 'gafgyt.tcp' in x:
        return 4
    elif 'gafgyt.udp' in x:
        return 5
    elif 'mirai.ack' in x:
        return 6
    elif 'mirai.scan' in x:
        return 7
    elif 'mirai.syn' in x:
        return 8
    elif 'mirai.udpplain' in x:
        return 9
    else :
        return 10

dfs = []
for x in list_of_files:
    if x in ['README.md', 'features.csv', 'device_info.csv', 'data_summary.csv', '.ipynb_checkpoints']:
        continue
    new_df = pd.read_csv('../input/nbaiot-dataset/' + x)
    new_df['type'] = type(x)
    dfs.append(new_df)
```

```python
allSamples = pd.concat(dfs, axis=0, ignore_index=True)
allSamples
```

```python
allSamples = allSamples.drop_duplicates()
allSamples.reset_index(inplace=True)
allSamples.shape
```

```python
allSamples.drop('index', inplace=True, axis=1)
allSamples
```

## Type dataframes

```python
typeFrames = []
```

```python
for i in range(0,11):
    typeFrames.append(allSamples[allSamples.type==i])
    
```

```python
for df in typeFrames:
    print(df.shape)
```

```python
li = []
i = 0
for df in typeFrames:
    if(i==2 or i==3):
        i = i+1
        li.append(df)
        print(df.shape)
        continue
    df = df.sample(n=62213)
    li.append(df)
    print(df.shape)
    i = i+1
```

```python
for df in li:
    print(df.shape)
```

```python
improvedType = pd.concat(li, axis=0, ignore_index=True)
improvedType.shape
```

```python
Tx = improvedType.drop('type', axis=1)
Ty = improvedType['type']
```

```python
Ty = pd.DataFrame(Ty)
```

```python
pca = PCA(n_components=32, whiten=True)
Tx = pca.fit_transform(StandardScaler().fit_transform(Tx))
```

```python
Tx.shape
```

```python
TrainTx, TestTx, TrainTy, TestTy = train_test_split(Tx, Ty, test_size=0.3, random_state=42)
TrainTx, ValTx, TrainTy, ValTy = train_test_split(TrainTx, TrainTy, test_size=0.3, random_state=42)
```

```python
TrainTx = pd.DataFrame(TrainTx)
TrainTy = pd.DataFrame(TrainTy)
ValTx = pd.DataFrame(ValTx)
ValTy = pd.DataFrame(ValTy)
TestTx = pd.DataFrame(TestTx)
TestTy = pd.DataFrame(TestTy)
```

```python
TrainTx.to_csv('TrainTx.csv', index=False)
TrainTy.to_csv('TrainTy.csv', index=False)
ValTx.to_csv('ValTx.csv', index=False)
ValTy.to_csv('ValTy.csv', index=False)
TestTx.to_csv('TestTx.csv', index=False)
TestTy.to_csv('TestTy.csv', index=False)
```
